üßπ Data Cleaning Step 2: Handling Missing Values


```python
# First, let's look at what's in our altitude column
print("Altitude column info:")
print(coffee_clean['altitude'].dtype)
print("\nSample of altitude values:")
print(coffee_clean['altitude'].head(10))

# Check if there are any non-numeric values
print("\nUnique values in altitude column:")
print(coffee_clean['altitude'].unique()[:10])
```

```python
# Clean altitude column
def clean_altitude(alt):
    if isinstance(alt, str):
        # Extract numbers from strings like "1200m" or "1200-1500"
        import re
        numbers = re.findall(r'\d+', str(alt))
        if numbers:
            # If range is given, take the average
            return sum(int(num) for num in numbers) / len(numbers)
    elif pd.notnull(alt):
        return float(alt)
    return np.nan

# Clean the altitude column
coffee_clean['altitude'] = coffee_clean['altitude'].apply(clean_altitude)

# Now we can fill missing values with median
coffee_clean['altitude'] = coffee_clean['altitude'].fillna(coffee_clean['altitude'].median())

# Verify the results
print("\nAltitude statistics after cleaning:")
print(coffee_clean['altitude'].describe())
```

```python
# For text columns, we'll fill with 'Unknown'
text_columns = ['farm_name', 'mill', 'region', 'producer']
for col in text_columns:
    coffee_clean[col] = coffee_clean[col].fillna('Unknown')
```

üßπ Data Cleaning Step 3: Detecting Outliers

```python
print("\nüîç Let's look for outliers in our numerical scores")
```

```python
# Create a box plot for all scoring categories
plt.figure(figsize=(15, 6))
score_columns = [col for col in coffee_clean.columns if col.endswith('_score')]
coffee_clean[score_columns].boxplot()
plt.xticks(rotation=45)
plt.title('Distribution of Coffee Scores')
plt.ylabel('Score')
plt.tight_layout()
plt.show()
```

```python
# Calculate outlier boundaries for total score
Q1 = coffee_clean['total_score'].quantile(0.25)
Q3 = coffee_clean['total_score'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
print(f"\nüìä Total Score Statistics:")
print(f"Lower bound for outliers: {lower_bound:.2f}")
print(f"Upper bound for outliers: {upper_bound:.2f}")
print(f"Number of outliers below: {len(coffee_clean[coffee_clean['total_score'] < lower_bound])}")
print(f"Number of outliers above: {len(coffee_clean[coffee_clean['total_score'] > upper_bound])}")
```
